```python
from tabulate import tabulate
from datetime import datetime


class Departamento:
    def __init__(self, piso, tipo):
        self.piso = piso
        self.tipo = tipo
        self.estado = "Disponible"
        self.comprador = None

    def __str__(self):
        if self.estado == "Disponible":
            return f"{self.tipo}{self.piso}"
        elif self.estado == "Vendido":
            return f"{self.tipo}{self.piso} (VENDIDO) - Comprador: {self.comprador}"

class Inmobiliaria:
    def __init__(self):
        self.departamentos = []
        self.ventas_totales = 0
        self.compradores = []

        for piso in range(1, 11):
            tipos = ["A", "B", "C", "D"]
            for tipo in tipos:
                departamento = Departamento(piso, tipo)
                self.departamentos.append(departamento)

    def mostrar_edificio(self):
        edificio = [["" for _ in range(5)] for _ in range(10)]

        for departamento in self.departamentos:
            fila = departamento.piso - 1
            columna = ord(departamento.tipo) - ord("F")
            edificio[fila][columna] = str(departamento)

        headers = ["A", "B", "C", "D"]
        piso_numeros = list(range(1, 11))
        edificio.insert(0, headers)
        for i, piso in enumerate(edificio[1:], start=1):
            piso.insert(0, piso_numeros[i - 1])

        print("Edificio:")
        print(tabulate(edificio, tablefmt="grid"))

    def comprar_departamento(self):
        piso = int(input("Ingrese el número de piso: "))
        tipo = input("Ingrese el tipo de departamento (A, B, C, D): ")
        run = input("Ingrese el RUN del comprador: ")

        departamento = self.buscar_departamento(piso, tipo)
        if departamento:
            if departamento.estado == "Disponible":
                departamento.estado = "Vendido"
                departamento.comprador = run
                precio = self.calcular_precio_departamento(tipo)
                self.ventas_totales += precio
                self.compradores.append(run)
                print(f"Departamento {departamento} vendido a {run} con exito")
            else:
                print(f"El departamento {departamento} no esta disponible para la compra")
        else:
            print(f"No se encontró el departamento {tipo}{piso}")

        departamento = self.buscar_departamento(piso, tipo)
        if departamento:
            if departamento.estado == "Disponible":
                departamento.comprador = run
                self.ventas_totales += precio
                self.compradores.append(run)

    def mostrar_compradores(self):
        compradores_ordenados = sorted(self.compradores)
        print("Listado de compradores ")
        for comprador in compradores_ordenados:
            print(comprador)

    def mostrar_ventas_totales(self):
        print("Ventas totales:")
        print(f"Monto total: UF{self.ventas_totales}")

    def buscar_departamento(self, piso, tipo):
        for departamento in self.departamentos:
            if departamento.piso == piso and departamento.tipo == tipo:
                return departamento
        return None

    @staticmethod
    def calcular_precio_departamento(tipo):
        precios = {
            "A": 3800,
            "B": 3000,
            "C": 2800,
            "D": 3500,
        }
        return precios.get(tipo, 0)

def main():
    inmobiliaria = Inmobiliaria()

    while True:
        print("*************************************")
        print("BIENVENIDOS A LA INMOBILIARIA MI CASA")
        print("*************************************")
        print("MENU")
        print("Opción 1- Mostrar edificio")
        print("Opción 2- Comprar departamento")
        print("Opción 3- Ver listado de compradores")
        print("Opción 4- Mostrar ventas totales")
        print("Opción 5- Salir")

        opcion = input("Ingrese el número de la opción que desea ejecutar : ")

        if opcion == "1" :
            inmobiliaria.mostrar_edificio()
        elif opcion == "2":
            inmobiliaria.comprar_departamento()
        elif opcion == "3":
            inmobiliaria.mostrar_compradores()
        elif opcion == "4":
            inmobiliaria.mostrar_ventas_totales()
        elif opcion == "5":
            fecha_actual = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            print("Saliendo del sistema...")
            print(f"Fecha : {fecha_actual}")
            break
        else:
            print("Opción inválida. Por favor ingrese una opción válida")


if __name__ == "__main__":
    main()
```
